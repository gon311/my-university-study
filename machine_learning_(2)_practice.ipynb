[1] 필요한 패키지 불러오기

```python
import numpy as np
```

[2] 선형회귀 클래스 선언

```python
class LinRegression() :
    def __init__(self, nDim, w=[]):
        self.nDim = nDim
        if len(w) > 0:
            self.w = np.array(w)
        else:
            self.w = np.random.rand(nDim+1)

    def fit(self, X, y, m, epochs, eta=0.01):
        cost = 0
        for i in range(m) :
            y_hat = np.dot(self.w, X[i])
            err = y_hat - y[i]
            cost += err * err
        cost /= m
        print("Epochs = {:4d}".format(0), end='')
        for i in range(self.nDim+1):
            print("   w{} = {:.4f}".format(i, self.w[i]), end='')
        print("   cost = {:.6f}".format(cost))

        for j in range(i, epochs+1):
            dw = np.zeros(self.nDim+1)
            for i in range(m):
                y_hat = np.dot(self.w, X[i])
                err = y_hat - y[i]
                dw += err * X[i]
            # 선형회귀 계수 업데이트
            self.w -= eta * 2.0 * dw * m

            # 20회 주기로 학습내용 출력
            if j % 20 == 0 :
                cost = 0
                for i in range(m) :
                    y_hat = np.dot(self.w, X[i])
                    err = y_hat - y[i]
                    cost += err * err
                cost /= m
                print("Epochs = {:4d}".format(0), end='')
                for i in range(self.nDim+1):
                    print("   w{} = {:.4f}".format(i, self.w[i]), end='')
                print("   cost = {:.6f}".format(cost))

    def predict(self, x) :
        return np.dot(self.w, x)
```

[3] 선형회귀를 위한 설정

```python
nSamples = 3
nDim = 1
nEpochs = 6000
eta = 0.01
```

[4] 독립변수와 종속변수 및 파라미터 초기화

```python
# x: 독립변수([1, x]), y : 종속변수
x = np.array([[1,1], [1,2], [1,3]])
y = np.array([2, 3, 4])

# 선형회귀의 계수
w = np.array([0.3, 0.5])
```

[5] 선형회귀 객체 생성 및 학습

```python
linReg = LinRegression(1, w)
linReg.fit(x, y, nSamples, nEpochs, 0.01)
```

[6] 학습된 선형회귀 모델을 이용한 예측  

```python
# x = 1.5 -> y_hat
xx = np.array([1, 1.5])
print("x = {:.2f} --> y_hat = {:.1f}".format(xx[1], linReg.predict(xx)))

# x = 4 -> y_hat
xx = np.array([1, 4])
print("x = {:.2f} --> y_hat = {:.1f}".format(xx[1], linReg.predict(xx)))
```

# 로지스틱 회귀 구현

[1] 로지스틱 함수 및 이진 교차 엔트로피 함수

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def binary_CE(y, y_hat):
    return -y * np.log(y_hat) - (1 - y) * np.log(1 - y_hat)
```

[2] 로지스틱 회귀 클래스 선언

```python
class LogisticRegression():
    def __init__(self, nDim, w=[]):
        self.nDim = nDim
        if len(w) > 0 :
            self.w = np.array(w)
        else:
            self.w = np.random.rand(nDim+1)

    def fit(self, X, y, m, epochs, eta=0.01) :
        cost = 0
        for i in range(m):
            y_hat = sigmoid(np.dot(self.w, X[i]))
            cost += binary_CE(y[i], y_hat)
        cost /= m
        print("Epochs = {:4d}".format(0), end='')
        for i in range(self.nDim+1):
            print("   w{} = {:.4f}".format(i, self.w[i]), end='')
        print("    cost = {:.6f}".format(cost))

        for j in range(1, epochs+1):
            dw = np.zeros(self.nDim+1)
            for i in range(m):
                y_hat = sigmoid(np.dot(self.w, x[i]))
                err = y_hat - y[i]
                dw += err * x[i]
            #로지스틱 회귀 계수 업데이트
            self.w -= eta * dw / m

            # 20회 주기로 학습내용 출력
            if j % 20 == 0 :
                cost = 0
                for i in range(m):
                    y_hat = sigmoid(np.dot(self.w, X[i]))
                    cost += binary_CE(y[i], y_hat)
                cost /= m
                print("Epochs = {:4d}".format(0), end='')
                for i in range(self.nDim+1):
                    print("   w{} = {:.4f}".format(i, self.w[i]), end='')
                print("    cost = {:.6f}".format(cost))

    def predict(self, x) :
        return sigmoid(np.dot(self.w, x))
```

[3] 로지스틱 회귀를 위한 설정

```python
nSamples = 6
nDim = 2
nEpochs = 8000
eta = 0.01
```

[4] 독립변수, 종속변수 및 파라미터 초기화

```python
# x : 독립변수([1, x1, x2]), y : 종속변수
x = np.array([[1, 1, 1], [1, 2, 1], [1, 1, 2],
              [1, 2, 3], [1, 3, 2], [1, 3, 3]])
y = np.array([0, 0, 0, 1, 1, 1])

# 로지스틱 회귀의 계수
w = np.array([0.5, -0.5, 1.0])
```

[5] 로지스틱 회귀 객체 생성 및 학습

```python
logisticReg = LogisticRegression(2, w)
logisticReg.fit(x, y, nSamples, nEpochs, 0.01)
```

[6] 학습된 선형회귀 모데을 이용한 예측

```python
# x = [2.5, 2.5] -> y_hat
xx = np.array([1, 2.5, 2.5])
print("x = ({:.2f}, {:.2f}) --> y_hat = {:.3f}".format(xx[1], xx[2], logisticReg.predict(xx)))

# x = [1.5, 1.5] -> y_hat
xx = np.array([1, 1.5, 1.5])
print("x = ({:.2f}, {:.2f}) --> y_hat = {:.3f}".format(xx[1], xx[2], logisticReg.predict(xx)))
```
